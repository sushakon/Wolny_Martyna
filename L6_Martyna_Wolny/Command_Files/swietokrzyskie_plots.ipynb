```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Set path to data
DIR = "../Analysis_Data/"
```

```python
data_vacuum = pd.read_csv(f"{DIR}data_vacuum.csv")
beko_data = pd.read_csv(f"{DIR}beko_data.csv")
tefal_data = pd.read_csv(f"{DIR}tefal_data.csv")
electrolux_data = pd.read_csv(f"{DIR}electrolux_data.csv")
samsung_data = pd.read_csv(f"{DIR}samsung_data.csv")
dyson_data = pd.read_csv(f"{DIR}dyson_data.csv")
data_client = pd.read_csv(f"{DIR}data_client.csv")
client_woman = pd.read_csv(f"{DIR}client_woman.csv")
client_man = pd.read_csv(f"{DIR}client_man.csv")
client_bd = pd.read_csv(f"{DIR}client_bd.csv")
```

```python
#Wykres liczby osób według płci
client_woman_count = client_woman.shape[0]
client_man_count = client_man.shape[0]  
client_bd_count = client_bd.shape[0]
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
legend = ['Kobieta', 'Mężczyzna', 'Brak danych']
gender = [client_woman_count, client_man_count, client_bd_count]
# ax.bar(legend, gender)
ax.bar_label(ax.bar(legend, gender))
plt.title('Wykres liczby osób danej płci')
plt.xlabel('Płeć')
plt.ylabel('Liczba')
plt.show()
```

```python
#Wykres średniego wieku osoby danej płci
client_woman_age_avg = client_woman['Age'].mean()
client_man_age_avg = client_man['Age'].mean()
client_bd_age_avg = client_bd['Age'].mean()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
legend = ['Kobieta', 'Mężczyzna', 'Brak danych']
gender = [client_woman_age_avg, client_man_age_avg, client_bd_age_avg]
# ax.bar(legend, gender)
ax.bar_label(ax.bar(legend, gender))
plt.title('Wykres średniego wieku osoby danej płci')
plt.xlabel('Płeć')
plt.ylabel('Średni wiek')
plt.show()
```

```python
#Wykres przebiegu zmiany średniej oceny w zależności od dni od zakupu
beko_rating = beko_data.drop(['Marka'], axis=1)
dyson_rating = dyson_data.drop(['Marka'], axis=1)
electrolux_rating = electrolux_data.drop(['Marka'], axis=1)
samsung_rating = samsung_data.drop(['Marka'], axis=1)
tefal_rating = tefal_data.drop(['Marka'], axis=1)
```

```python
beko_rating_avg = beko_rating.groupby(['Dni_od_zakupu'])['Ocena'].agg(lambda x: x.unique().sum()/x.nunique())
dyson_rating_avg = dyson_rating.groupby(['Dni_od_zakupu'])['Ocena'].agg(lambda x: x.unique().sum()/x.nunique())
electrolux_rating_avg = electrolux_rating.groupby(['Dni_od_zakupu'])['Ocena'].agg(lambda x: x.unique().sum()/x.nunique())
samsung_rating_avg = samsung_rating.groupby(['Dni_od_zakupu'])['Ocena'].agg(lambda x: x.unique().sum()/x.nunique())
tefal_rating_avg = tefal_rating.groupby(['Dni_od_zakupu'])['Ocena'].agg(lambda x: x.unique().sum()/x.nunique())
```

```python
beko_rating_avg.plot()
plt.title('Wykres przebiegu zmiany średniej oceny dla firmy Beko')
plt.xlabel('Dni od zakupu')
plt.ylabel('Średnia ocena')
plt.show()
```

```python
dyson_rating_avg.plot()
plt.title('Wykres przebiegu zmiany średniej oceny dla firmy Dyson')
plt.xlabel('Dni od zakupu')
plt.ylabel('Średnia ocena')
plt.show()
```

```python
electrolux_rating_avg.plot()
plt.title('Wykres przebiegu zmiany średniej oceny dla firmy Electrolux')
plt.xlabel('Dni od zakupu')
plt.ylabel('Średnia ocena')
plt.show()
```

```python
samsung_rating_avg.plot()
plt.title('Wykres przebiegu zmiany średniej oceny dla firmy Samsung')
plt.xlabel('Dni od zakupu')
plt.ylabel('Średnia ocena')
plt.show()
```

```python
tefal_rating_avg.plot()
plt.title('Wykres przebiegu zmiany średniej oceny dla firmy Tefal')
plt.xlabel('Dni od zakupu')
plt.ylabel('Średnia ocena')
plt.show()
```
