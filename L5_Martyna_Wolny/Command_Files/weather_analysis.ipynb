```python
import numpy as np
import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'
```

Ustawienie ścieżki do folderu Analysis_Data, który zawiera kopię danych do przetworzenia:

```python
# Set path to data
DIR = "../Analysis_Data/"
```

Tworzenie tytułów kolumn do Data Frame:

```python
columns = ['Id']
for i in range(1, 32):
    columns.append(str(i))
    
# print(columns)
```

Wczytanie danych do numpy array i edycja otrzymanych danych, aby uzyskać Data Frame:

```python
filename = f"{DIR}weather.txt"
data_original = np.loadtxt(filename, delimiter='\s+', dtype=str)
# print(data_original)
```

```python
data = []
for i in range (len(data_original)):
    data_original[i] = data_original[i].replace('I-9999', '')
    data_original[i] = data_original[i].replace('S-9999', '')
    data_original[i] = data_original[i].replace(' I ', '')
    data_original[i] = data_original[i].replace(' S ', '')
    data_original[i] = data_original[i].replace(' I', '')
    data_original[i] = data_original[i].replace(' S', '')
    # data_original[i] = data_original[i].replace(' ', ',')
    # print(i)
    
# print(data_original)
    
for i in range (len(data_original)):
    data_original[i].strip()
    data.append(data_original[i].split())

# print(data)
```

Tworzenie Data Frame zawierającej oryginalne dane po usunięciu niepotrzebnych separatorów:

```python
weather_data = pd.DataFrame(data)
weather_data.drop([32, 33], axis=1, inplace=True)
weather_data.columns = columns
weather_data
# weather_data.columns = columns
```

Zapis zedytowanych danych do pliku csv:

```python
weather_data.to_csv(path_or_buf=f"{DIR}weather_data.csv", index=False)
```

Wybór wierszy zawierających dane z roku 2010:

```python
weather_data_2010 = weather_data[weather_data.Id.str.contains('2010', case=False)]
weather_data_2010.to_csv(path_or_buf=f"{DIR}weather_data_2010.csv", index=False)
# weather_data_2010
```

Zamiana błędnych danych o wartości -9999 lub pustych danych None na numpy.nan:

```python
weather_data_2010 = weather_data_2010.to_numpy()


for i in range (len(weather_data_2010)):
    for j in range(len(weather_data_2010[i])):
        if weather_data_2010[i][j] == '-9999' or weather_data_2010[i][j] == None:
            weather_data_2010[i][j] = np.nan
            
            
# print(min_values_data)

weather_data_2010 = pd.DataFrame(weather_data_2010)
weather_data_2010.columns = columns
weather_data_2010
```

Rozdział danych i wyznaczenie wartości minimalnej i maksymalnej temperatury w miesiącu oraz średnich opadów:

```python
min_values_data = weather_data_2010[weather_data_2010.Id.str.contains('MIN', case=False)]
max_values_data = weather_data_2010[weather_data_2010.Id.str.contains('MAX', case=False)]
prcp_values_data = weather_data_2010[weather_data_2010.Id.str.contains('PRCP', case=False)]
```

```python
min_values_data = min_values_data.set_index('Id').transpose()
max_values_data = max_values_data.set_index('Id').transpose()
prcp_values_data = prcp_values_data.set_index('Id').transpose()
```

```python
for col in min_values_data:
    min_values_data[col] = min_values_data[col].astype('Int64')
    
for col in max_values_data:
    max_values_data[col] = max_values_data[col].astype('Int64')
    
for col in prcp_values_data:
    prcp_values_data[col] = prcp_values_data[col].astype('Int64')
```

```python
min_values = min_values_data.min(skipna=True)
# min_values
```

```python
min_values = min_values.to_frame('MIN TEMP').reset_index()
min_values
```

```python
max_values = max_values_data.min(skipna=True)
# max_values
```

```python
max_values = max_values.to_frame('MAX TEMP').reset_index()
max_values
```

```python
prcp_average = prcp_values_data.mean(skipna=True)
# prcp_average
```

```python
prcp_average = prcp_average.to_frame('AVERAGE PRCP').reset_index()
prcp_average
```

Utworzenie bazy zawierającej informacje o minimalnej i maksymalnej temperaturze w miesiącu oraz średnich opadach w roku 2010 w pierwszych pięciu miesiącach:

```python
station_id = []
date = []

for i in range (1, 6):
    if i < 10:
        date.append('2010-0{month}'.format(month = i))
        station_id.append('MMX17004')
```

```python
min_values_5 = min_values.head(5).drop('Id', axis=1)
max_values_5 = max_values.head(5).drop('Id', axis=1)
prcp_average_5 = prcp_average.head(5).drop('Id', axis=1)
```

```python
min_values_5.insert(0, 'Date', date, True)
max_values_5.insert(0, 'Date', date, True)
prcp_average_5.insert(0, 'Date', date, True)
```

```python
weather_data_2010_5_months = min_values_5
weather_data_2010_5_months = weather_data_2010_5_months.join(max_values_5.set_index('Date'), on='Date')
weather_data_2010_5_months = weather_data_2010_5_months.join(prcp_average_5.set_index('Date'), on='Date')
weather_data_2010_5_months.insert(0, 'Id', station_id, True)
weather_data_2010_5_months
```

Zapisanie otrzymanej bazy do pliku csv:

```python
weather_data_2010_5_months.to_csv(path_or_buf=f"{DIR}weather_data_2010_5_months.csv", index=False)
```
